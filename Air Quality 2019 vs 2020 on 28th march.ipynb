```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
```

```python
AQI = pd.read_csv('MARCH_AQI.csv')
```

```python
AQI
```

```python
type(AQI)
```

```python
Y2019 = AQI[AQI.Year == 'Y2019']
```

```python
Y2019
```

```python
plt.plot(Y2019.Cities, Y2019.AQI_Morn, Y2019.AQI_Eve)
plt.xlabel("Cities")
plt.ylabel("AQI")
plt.legend(["AQI at 10 AM", "AQI at 6 PM"])
plt.show()
```
